# **Classifying Columns Based On Data Behaviour**

```python
import pandas as pd
import numpy as np

df = pd.read_csv("Py_Raw_Cleaned_v2.csv")

# Define thresholds and grouping logic

sparse_cols = [col for col in df.columns if (df[col] == 0).mean() > 0.8]

ratio_cols = [col for col in df.columns
              if df[col].max() <= 10 and df[col].min() >= 0
              and df[col].nunique() > 2
              and col not in sparse_cols]

continuous_cols = list(set(df.columns) - set(sparse_cols) - set(ratio_cols))
df.shape

```

```python
print("Continuous columns:", len(continuous_cols))
print(continuous_cols[:10])   # sample
print("\nRatio/bounded columns:", len(ratio_cols))
print(ratio_cols)
print("\nSparse columns:", len(sparse_cols))
print(sparse_cols)
df.shape

```

```python
feature_map = {
    "continuous": continuous_cols,
    "ratio": ratio_cols,
    "sparse": sparse_cols,
}

# Convert to a DataFrame for easy inspection or export
feature_map_df = (
    pd.DataFrame([(col, group) for group, cols in feature_map.items() for col in cols],
                 columns=['column_name', 'feature_type'])
)
feature_map_df.to_csv("feature_groups.csv", index=False)
df.shape

```

# **Handle Missing and Zero Values in Ratio Columns**

```python
# Replace zeros with NaN in ratio columns only
df[ratio_cols] = df[ratio_cols].replace(0, np.nan)

# Fill missing numeric values with median (robust)
df = df.fillna(df.median())
df.shape
```

# **Normalizing RATIO COLUMS using LOG TRANSFORMATION**

```python
for col in ratio_cols:
    if df[col].skew() > 1:
        df[col] = np.log1p(df[col])  # log1p(x) = log(1+x) avoids log(0)
df.shape
```

# **Aggregating the Sparse Columns into RARE CONSTRUCTS**

```python
df['rare_constructs'] = df[sparse_cols].sum(axis=1)
df = df.drop(columns=sparse_cols)
df.shape
```

# **Cap Outliers in Continuous Features**

```python
Q1 = df[continuous_cols].quantile(0.25)
Q3 = df[continuous_cols].quantile(0.75)
IQR = Q3 - Q1

def clip_outliers(col):
    lower = Q1[col.name] - 1.5 * IQR[col.name]
    upper = Q3[col.name] + 1.5 * IQR[col.name]
    return col.clip(lower, upper)

df[continuous_cols] = df[continuous_cols].apply(clip_outliers, axis=0)
df.shape
```

# **Remove MultiCollinearity**

```python
corr_matrix = df[continuous_cols + ratio_cols].corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.9)]
df = df.drop(columns=to_drop)

print("Dropped highly correlated features:", to_drop)
df.shape
```

# **Feature Scaling**

```python
# ✅ Ensure expected columns exist
expected_cols = continuous_cols + ratio_cols
missing_cols = [c for c in expected_cols if c not in df.columns]
print("❌ Missing columns:", missing_cols if missing_cols else "None ✅")

# ✅ Proceed only with columns that exist in df
scaler_cols = [c for c in expected_cols if c in df.columns]

# ✅ Apply scaling
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Standard Scaler version
df_lr = df.copy()
scaler_std = StandardScaler()
df_lr[scaler_cols] = scaler_std.fit_transform(df_lr[scaler_cols])

# MinMax Scaler version
df_nn = df.copy()
scaler_mm = MinMaxScaler()
df_nn[scaler_cols] = scaler_mm.fit_transform(df_nn[scaler_cols])

print(f"✅ Scaled {len(scaler_cols)} columns successfully!")
print(f"📊 df_lr shape: {df_lr.shape}")
print(f"📊 df_nn shape: {df_nn.shape}")
df.shape

```

# **Train Test Split**

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load your preprocessed features
df = pd.read_csv("preprocessed_dataset.csv")

# Assuming your target is already loaded as a pandas Series named 'target'
print("Target preview:")
print(target.head())

# Safety check
assert len(df) == len(target), f"❌ Mismatch: features={len(df)} vs target={len(target)}"

# Split the data (for regression → no stratify)
X_train, X_test, y_train, y_test = train_test_split(
    df,
    target,
    test_size=0.2,       # 20% test data
    random_state=42,     # reproducible split
    shuffle=True
)

# Save all splits as CSV files
X_train.to_csv("X_train.csv", index=False)
X_test.to_csv("X_test.csv", index=False)
y_train.to_csv("y_train.csv", index=False)
y_test.to_csv("y_test.csv", index=False)

print("\n✅ Data successfully split and saved!")
print("📊 X_train:", X_train.shape)
print("📊 X_test:", X_test.shape)
print("📈 y_train:", y_train.shape)
print("📉 y_test:", y_test.shape)
print("\nFiles saved as: X_train.csv, X_test.csv, y_train.csv, y_test.csv")
```

# **Linear Regression**

```python
from sklearn.linear_model import LinearRegression

# Initialize model
lr_model = LinearRegression()

# Train it
lr_model.fit(X_train, y_train)

print("✅ Model training complete!")

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Predict on test data
y_pred = lr_model.predict(X_test)

# Evaluate
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"📈 R² Score: {r2:.4f}")
print(f"📉 MAE: {mae:.4f}")
print(f"📉 MSE: {mse:.4f}")
print(f"📉 RMSE: {rmse:.4f}")

```

# *Random Forest*

```python
# =====================================================
# ⚡ FAST RANDOM FOREST REGRESSION PIPELINE (RandomizedSearchCV)
# =====================================================

import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
import joblib

# -----------------------------------------------------
# 1️⃣ Load pre-split datasets
# -----------------------------------------------------
X_train = pd.read_csv("X_train.csv")
X_test = pd.read_csv("X_test.csv")
y_train = pd.read_csv("y_train.csv").squeeze()  # convert DataFrame -> Series
y_test = pd.read_csv("y_test.csv").squeeze()

print("✅ Data loaded successfully!")
print(f"X_train: {X_train.shape}, X_test: {X_test.shape}")
print(f"y_train: {y_train.shape}, y_test: {y_test.shape}")

# -----------------------------------------------------
# 2️⃣ Hyperparameter tuning with RandomizedSearchCV
# -----------------------------------------------------
param_dist = {
    'n_estimators': [50, 100, 150, 200],   # fewer trees for speed
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

rf = RandomForestRegressor(random_state=42)

random_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_dist,
    n_iter=20,          # only 20 random combos
    cv=2,               # fewer folds for speed
    scoring='r2',
    n_jobs=-1,
    verbose=1,
    random_state=42
)

print("\n🔍 Running RandomizedSearchCV for fast hyperparameter tuning...")
random_search.fit(X_train, y_train)

best_rf = random_search.best_estimator_

print("\n🎯 Best Parameters Found:")
print(random_search.best_params_)

# -----------------------------------------------------
# 3️⃣ Model evaluation
# -----------------------------------------------------
y_pred = best_rf.predict(X_test)

r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("\n📊 Model Evaluation Metrics:")
print(f"R² Score: {r2:.4f}")
print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")

# -----------------------------------------------------
# 4️⃣ Feature importance visualization
# -----------------------------------------------------
importances = best_rf.feature_importances_
features = X_train.columns

imp_df = pd.DataFrame({'Feature': features, 'Importance': importances})
imp_df = imp_df.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10,6))
plt.barh(imp_df['Feature'][:15], imp_df['Importance'][:15])
plt.gca().invert_yaxis()
plt.title("Top 15 Important Features")
plt.xlabel("Feature Importance")
plt.show()

# -----------------------------------------------------
# 5️⃣ Actual vs Predicted plot
# -----------------------------------------------------
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred, alpha=0.6)
plt.xlabel("Actual Energy Consumption (Joules)")
plt.ylabel("Predicted Energy Consumption (Joules)")
plt.title(f"Random Forest: Actual vs Predicted (R² = {r2:.3f})")
plt.grid(True)
plt.show()

# -----------------------------------------------------
# 6️⃣ Save the trained model
# -----------------------------------------------------
joblib.dump(best_rf, "random_forest_fast_model.pkl")
print("\n💾 Tuned Random Forest model saved as 'random_forest_fast_model.pkl'")
```

# **XGBoost**

```python
# =====================================================
# ⚡ XGBoost Regression Pipeline
# =====================================================

import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
import joblib

# -----------------------------------------------------
# 1️⃣ Load pre-split datasets
# -----------------------------------------------------
X_train = pd.read_csv("X_train.csv")
X_test = pd.read_csv("X_test.csv")
y_train = pd.read_csv("y_train.csv").squeeze()
y_test = pd.read_csv("y_test.csv").squeeze()

print("✅ Data loaded successfully!")
print(f"X_train: {X_train.shape}, X_test: {X_test.shape}")
print(f"y_train: {y_train.shape}, y_test: {y_test.shape}")

# -----------------------------------------------------
# 2️⃣ Define parameter distribution for RandomizedSearchCV
# -----------------------------------------------------
param_dist = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7, 10],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0]
}

xgb = XGBRegressor(random_state=42, objective='reg:squarederror', n_jobs=-1)

random_search = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=param_dist,
    n_iter=20,          # try 20 random combinations
    cv=2,               # 2-fold CV for speed
    scoring='r2',
    n_jobs=-1,
    verbose=1,
    random_state=42
)

print("\n🔍 Running RandomizedSearchCV for XGBoost...")
random_search.fit(X_train, y_train)

best_xgb = random_search.best_estimator_

print("\n🎯 Best Parameters Found:")
print(random_search.best_params_)

# -----------------------------------------------------
# 3️⃣ Model evaluation
# -----------------------------------------------------
y_pred = best_xgb.predict(X_test)

r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("\n📊 XGBoost Model Evaluation Metrics:")
print(f"R² Score: {r2:.4f}")
print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")

# -----------------------------------------------------
# 4️⃣ Feature importance visualization
# -----------------------------------------------------
importances = best_xgb.feature_importances_
features = X_train.columns

imp_df = pd.DataFrame({'Feature': features, 'Importance': importances})
imp_df = imp_df.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10,6))
plt.barh(imp_df['Feature'][:15], imp_df['Importance'][:15])
plt.gca().invert_yaxis()
plt.title("Top 15 Important Features (XGBoost)")
plt.xlabel("Feature Importance")
plt.show()

# -----------------------------------------------------
# 5️⃣ Actual vs Predicted plot
# -----------------------------------------------------
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred, alpha=0.6)
plt.xlabel("Actual Energy Consumption (Joules)")
plt.ylabel("Predicted Energy Consumption (Joules)")
plt.title(f"XGBoost: Actual vs Predicted (R² = {r2:.3f})")
plt.grid(True)
plt.show()

# -----------------------------------------------------
# 6️⃣ Save the trained model
# -----------------------------------------------------
joblib.dump(best_xgb, "xgboost_model.pkl")
print("\n💾 XGBoost model saved as 'xgboost_model.pkl'")
```
